```python
import pymongo

from string import punctuation

import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
scrapy = client['scrapy']
item_info = scrapy['item_infoX']
```

```python
for i in item_info.find().limit(300):
    if i['area']:
        area = [ i for i in i['area'] if i not in punctuation]
    else:
        area = [ '不明' ]
    print(area)
```

```python
for i in item_info.find():
    if i['area']:
        area = [ i for i in i['area'] if i not in punctuation]
    else:
        area = [ '不明' ]
    # item_info.update_one({'_id': i['_id']} , {'$set': {'area': area}})
```

```python
area_list = []
for i in item_info.find():
    area_list.append(i['area'][0])
area_index = list(set(area_list))
print(area_index)
```

```python
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(post_times)
```

```python
def generator(types):
    for area, times in zip(area_index, post_times):
        data = {
            'name': area,
            'data': [times],
            'type': types
        }
        yield data
```

```python
for i in generator('column'):
    print(i)
```

```python
options = dict(
    title=dict(text='Zoom me!'),
    chart=dict(zoomType='xy')
)

series2 = [ i for i in generator('column') ]


charts.plot(series2, options=options, show='inline')
```
