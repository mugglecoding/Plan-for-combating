```python
from __future__ import print_function

import pymongo

import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
scrapy = client['scrapy']
item_info = scrapy['item_infoX']
```

```python
for i in item_info.find().limit(300):
    if len(i['cates']) > 3:
        print(i['cates'][2])
```

```python
cate_list = []
for i in item_info.find():
    cate_list.append(i['cates'][2])
cate_index = list(set(cate_list))
print(cate_index)
```

```python
post_times = []
for index in cate_index:
    post_times.append(cate_list.count(index))
print(post_times)
```

```python
def generator(types):
    for cate, times in zip(cate_list, post_times):
        data = {
            'name': cate,
            'data': [times],
            'type': types
        }
        yield data
```

```python
for i in generator('column'):
    print(i)
```

```python
options = dict(
    title=dict(text='Zoom me!'),
    chart=dict(zoomType='xy')
)

series2 = [ i for i in generator('column') ]


charts.plot(series2, options=options, show='inline')
```
