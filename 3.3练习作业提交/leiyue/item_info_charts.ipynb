```python
from __future__ import print_function

import pymongo
from datetime import datetime, timedelta
import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
scrapy = client['scrapy']
item_info = scrapy['item_infoX']
```

```python
for i in item_info.find({},{'pub_date': 1, '_id': 0}).limit(300):
    print(i)
```

```python
def get_all_date(date1, date2):
    the_date = datetime.strptime(date1, '%Y-%m-%d')
    end_date = datetime.strptime(date2, '%Y-%m-%d')
    days = timedelta(days=1)
    while the_date <= end_date:
        yield the_date.date().strftime('%Y-%m-%d')
        the_date += days
```

```python
print([i for i in get_all_date('2016-1-27', '2016-1-28') ])
```

```python
def get_data_within(date1, date2, areas):
    for area in areas:
        area_day_posts = []
        for date in get_all_date(date1, date2):
            each_day_posts = item_info.count({'pub_date': date, 'area': area})
            area_day_posts.append(each_day_posts)
        data = {
            'name': area,
            'data': area_day_posts,
            'type': 'line' 
        }
        yield data
```

```python
for i in get_data_within('2015-12-24', '2016-01-05', [ u'' ]):
    print(i)
```

```python
area_list = []
for i in item_info.find():
    area_list.append(i['area'][0])
area_index = list(set(area_list))
print(area_index)
```

```python
start_date = '2015-12-24'
end_date = '2016-01-05'

options = dict(
    title=dict(text=''),
    chart=dict(zoomType='xy'),
    xAxis=dict(title=dict(text=''), categories=[ i for i in get_all_date(start_date, end_date) ]),
    yAxis=dict(title=dict(text=''))
)

series = [{
    'type': 'column',
    'name': 'Jane',
    'data': [3, 2, 1, 3, 4]
}, {
    'type': 'line',
    'name': 'John',
    'data': [2, 3, 5, 7, 6]
}, {
    'type': 'column',
    'name': 'Joe',
    'data': [4, 3, 3, 9, 0]
}, {
    'type': 'spline',
    'name': 'Average',
    'data': [3, 2.67, 3, 6.33, 3.33],
    'marker': {
        'lineWidth': 2,
        'lineColor': '#90ed7d',
        'fillColor': 'white'
    }
}]

series2 = [ i for i in get_data_within(start_date, end_date, area_index) ]

charts.plot(series2, options=options, show='inline')
```
