```python
from __future__ import print_function

import pymongo
from datetime import datetime, timedelta
import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
scrapy = client['scrapy']
item_info = scrapy['item_infoX']
```

```python
for i in item_info.find().limit(300):
    if len(i['cates']) > 3:
        print(i['cates'][2])
```

```python
cate_list = []
for i in item_info.find():
    cate_list.append(i['cates'][2])
cate_index = list(set(cate_list))
print(cate_index)
```

```python
def get_all_date(date1, date2):
    the_date = datetime.strptime(date1, '%Y-%m-%d')
    end_date = datetime.strptime(date2, '%Y-%m-%d')
    days = timedelta(days=1)
    while the_date <= end_date:
        yield the_date.date().strftime('%Y-%m-%d')
        the_date += days
```

```python
def get_data_within(date1, date2, cates):
    for cate in cates:
        cate_days_posts = []
        for date in get_all_date(date1, date2):
            each_day_posts = len(list(item_info.find({'pub_date': date, 'cates': cate},{})))
            cate_days_posts.append(each_day_posts)
        data = {
            'name': cate,
            'data': cate_days_posts,
            'type': 'line' 
        }
        yield data
```

```python
start_date = '2015-12-24'
end_date = '2016-01-05'

options = dict(
    title=dict(text='发帖量类目统计'),
    chart=dict(zoomType='xy'),
    xAxis=dict(title=dict(text='日期'), categories=[ i for i in get_all_date(start_date, end_date) ]),
    yAxis=dict(title=dict(text='数量'))
)

series = [{
    'type': 'column',
    'name': 'Jane',
    'data': [3, 2, 1, 3, 4]
}, {
    'type': 'line',
    'name': 'John',
    'data': [2, 3, 5, 7, 6]
}, {
    'type': 'column',
    'name': 'Joe',
    'data': [4, 3, 3, 9, 0]
}, {
    'type': 'spline',
    'name': 'Average',
    'data': [3, 2.67, 3, 6.33, 3.33],
    'marker': {
        'lineWidth': 2,
        'lineColor': '#90ed7d',
        'fillColor': 'white'
    }
}]

series2 = [ i for i in get_data_within(start_date, end_date, cate_index) ]

charts.plot(series2, options=options, show='inline')
```
